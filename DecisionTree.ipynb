```python
# imported the car data into a Pandas DataFrame
import pandas as pd
data = pd.read_csv(r'/content/cardekho_data.csv')
```

```python
# Display the first few rows of the DataFrame
print(data.head())
```

```python
# Check for missing values in each column
print(data.isnull().sum())
```

```python
# Display the data types of each column
print(data.dtypes)
```

```python
# Display summary statistics of numerical columns
print(data.describe())
```

```python
#LabelEncoder from sklearn.preprocessing is a tool for
#converting categorical labels into numeric format by
# assigning a unique integer to each label
from sklearn.preprocessing import LabelEncoder
```

```python
LabelEncoder = LabelEncoder()
```

```python
data['Fuel_Type']=LabelEncoder.fit_transform(data['Fuel_Type'])
data['Seller_Type']=LabelEncoder.fit_transform(data['Seller_Type'])
data['Transmission']=LabelEncoder.fit_transform(data['Transmission'])
```

```python
data.head()
```

```python
Features_cols = ['Year','Fuel_Type','Seller_Type','Transmission','Kms_Driven','Present_Price','Selling_Price']
```

```python
X = data[Features_cols]
Y = data.Owner
```

```python
#train_test_split from sklearn.model_selection
#splits a dataset into training and testing sets
# for machine learning,
# facilitating model evaluation on unseen data.
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Classifier = DecisionTreeClassifier(criterion='entropy')
```

```python

Classifier.fit(X_train,y_train)
```

```python
 Classifier.predict(X_test)
```

```python
Classifier.score(X_test,y_test)
```
